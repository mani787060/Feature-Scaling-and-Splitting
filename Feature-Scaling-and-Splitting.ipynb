```python
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('Titanic-Dataset.csv')[['Age', 'Pclass', 'SibSp', 'Parch', 'Survived']]
```

```python
df.head()  

# here, we have 4 input columns and 1 output column(i.e, 'Survived').
```

```python
# as we know, here in 'Age', there is some missing data, so we drop it

df.dropna(inplace=True)
```

```python
df.head()
```

```python
# so, we now separate our input and output columns

X = df.iloc[:,0:4]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
# checking for accuracy using cross_val_score

np.mean(cross_val_score(LogisticRegression(), X,y,scoring='accuracy', cv=20))
```

<h2> Applying Feature Construction </h2>

```python
# now, I just want to  make a new column by adding 'SibSp' and 'Parch' to 'Family_size'

X['Family_size'] = X['SibSp'] + X['Parch'] + 1  # we add '1' here to also include that person in this 
X.head()
```

```python
# now, I just make a function in which when you provide 'Family_size' then it returns you a value conditionally

def myfunc(num):
    if num==1:
        # alone 
        return 0
    elif num > 1 and num <=4:
        # small family
        return 1
    else:
         #large family
        return 2
```

```python
myfunc(4)
```

```python
# Add these new column to the main column

X['Family_type'] = X['Family_size'].apply(myfunc)
```

```python
X.head()
```

```python
X.drop(columns=['SibSp', 'Parch', 'Family_size'], inplace=True)
```

```python
X.head()
```

```python
# Again, apply cross_val_score to check accuracy after applying "Feature_Construction"

np.mean(cross_val_score(LogisticRegression(), X,y,scoring='accuracy', cv=20))

# so it's clear, After implementing "Feature Construction", Accuracy is increased little
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df['Name']

# So, it is written in the order of (lastName, surName fisrtName)  
```

```python
# But, we just want that Surname(salutation) from here, so usko 'Name' column se bahar nikalna pdega

# df['Name'].str.split(', ', expand=True)  --> so, this gives both comma ke baad wali chiz + comma ke phli waali chiz
# df['Name'].str.split(', ', expand=True)[1]  --> this is for comma ke baad wali chiz k liye (b/c sirName hmara whi pe hia, and hme usse find krna hai)


df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True) #--> we apply this and said that ki 'hme dot ke upr split krke dikhao'

# now, It is divided into part by part, So, we can easily extract whatever we want from here
```

```python
df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0] 

# Using this, I find all Surname(Salutation) from the 'Name' column
```

```python
df[['Title', 'Name']].head()
```

```python
# (df.groupby('Title').mean()['Survived']).sort_values(ascending=False) --> it gives an error, so use it 

(df.groupby('Title')['Survived'].mean()).sort_values(ascending=False)

# so here, you can check how the 'Title' related to 'Survived' column 
```

```python
# we again construct a new column 

df['Is_Married'] = 0
df['Is_Married'].loc[df['Title'] == 'Mrs'] = 1
```
